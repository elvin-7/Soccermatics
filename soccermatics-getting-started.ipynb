# Sign test

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import os
import pathlib
import warnings

pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

```python
train = pd.read_json("/kaggle/input/soccermatics/events/events_England.json")

player_df = pd.read_json("/kaggle/input/soccermatics/players.json")
```

Here, we are trying to figure out if Son Heung-Min shoots with both feet the same number of times using the sign test.

```python
shots = train[train['subEventName']=='Shot']
son_id = player_df[player_df['shortName']=='Son Heung-Min']['wyId'].iloc[0]

son_shots = shots[shots['playerId']==son_id]

lefty_shots = son_shots[son_shots.apply(lambda x:{'id':401}in x.tags,axis=1)]

righty_shots = son_shots[son_shots.apply(lambda x:{'id':402}in x.tags,axis=1)]

l = [1]*len(lefty_shots)
l.extend([-1]*len(righty_shots))
l
```

We now conduct a sign test to confirm whether or not son is ambidextrous. Our significance level will be set at 0.05.

```python
from statsmodels.stats.descriptivestats import sign_test
test = sign_test(l, mu0=0)
pvalue = test[1]

if pvalue < 0.05:
    print("P-value amounts to",str(pvalue)[:5],"- We reject null hypothesis - Heung-Min Son is not ambidextrous")
else:
    print("P-value amounts to",str(pvalue)[:5],"- We do not reject null hypothesis - Heung-Min Son is ambidextrous")
```

```python
print("Son has taken ",str(len(lefty_shots))," shots with his left foot")
print("And ",str(len(righty_shots))," shots with his right")
```

# Using t-tests

Here, we would be demonstrating how to check if values are significantly different from each other usin t-tests

```python
teams_df = pd.read_json("/kaggle/input/soccermatics/teams.json")
teams_df = teams_df.rename(columns = {'wyId':'teamId'})
teams_df.head()
```

```python
#get corners
corners = train[train["subEventName"]=="Corner"]
#count corners by team
corners_by_teams = corners.groupby(['teamId']).size().reset_index(name='counts')
#merge with team name
summary = corners_by_teams.merge(teams_df[["name","teamId"]],how="left",on=["teamId"])
#count corners by team by game
corners_by_game = corners.groupby(['teamId','matchId']).size().reset_index(name='counts')
#merge with team name
summary2 = corners_by_game.merge(teams_df[["name","teamId"]],how="left",on=["teamId"])
```

## One-sample one-sided t-test

We assume that it is established typically a team gets 6 corners in a match. Arsenal is an attacking team, we might think they get more corners than this.

```python
team_name = 'Arsenal'
ars_corners = summary2[summary2["name"]=='Arsenal']["counts"]

fig,ax=plt.subplots(1,1)
ax.hist(ars_corners,np.arange(0.01,20.5,1),color='red',edgecolor='white',linestyle='-',alpha=0.5,label=team_name,density=True,align='right')
ax.legend(loc='upper left')
ax.set_ylim(0,0.25)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('Corners')
ax.set_ylabel('Proportion of games')
ax.set_xticks(np.arange(0,21,step=1))
plt.show()

mean = ars_corners.mean()
std = ars_corners.std()

print("Arsenal typically had %.2f plus/minus %.2f corners per match in the 2017/18 season."%(mean,std))
```

Now, we use a one-sided t-test to check if Arsenal tok more corners than we might expect due to normal variation in the number of corners we tend to see. The significance level is set at 0.05

```python
from scipy.stats import ttest_1samp
t, pvaue = ttest_1samp(ars_corners,popmean=6)

print("The t-statistic is %.2f and the P-value is %.2f."%(t,pvalue))
if pvalue < 0.05:
    print("We reject null hypothesis - "+ team_name+" typically take more than 6 corners per match.")
else:
    print("We cannot reject null hypothesis - "+ team_name+" do not typically take more than 6 corners per match.")
```

## Two-sample two-sided t-test

```python
spurs_corners = summary2[summary2["name"]=="Tottenham Hotspur"]["counts"]
manu_corners = summary2[summary2["name"]=="Manchester United"]["counts"]

mean = spurs_corners.mean()
std = spurs_corners.std()
print('Spurs typically had %.2f plus/minus %.2f corners per match in the 2017/18 season.'%(mean,std))
std_error=std/np.sqrt(len(spurs_corners))
print('The standard error in the number of corners per match is %.4f'%std_error)

print()

mean = manu_corners.mean()
std = manu_corners.std()
print('United typically had %.2f plus/minus %.2f corners per match in the 2017/18 season.'%(mean,std))
std_error=std/np.sqrt(len(spurs_corners))
print('The standard error in the number of corners per match is %.4f'%std_error)
```

```python
def FormatFigure(ax):
    ax.legend(loc='upper left')
    ax.set_ylim(0,0.25)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylabel('')
    ax.set_xlabel('Corners')
    ax.set_ylabel('Proportion of games')
    ax.set_xticks(np.arange(0,21,step=1))


fig,ax=plt.subplots(1,1)
ax.hist(manu_corners, np.arange(0.01,15.5,1), color='red', edgecolor = 'red',linestyle='-',alpha=1.0, label="United", density=True,align='right')
ax.hist(spurs_corners, np.arange(0.01,15.5,1), alpha=0.25, color='blue', edgecolor = 'white', label='Spurs',  density=True,align='right')
FormatFigure(ax)
```

```python
# Here we test if United had a different average corners per game than Spurs.
# We set the significance level at 0.05.


from scipy.stats import ttest_ind
t, pvalue  = ttest_ind(a=manu_corners, b=spurs_corners, equal_var=True)

print("The t-staistic is %.2f and the P-value is %.2f."%(t,pvalue))
if pvalue < 0.05:
    print("We reject null hypothesis - United took different number of corners per game than Spurs")
else:
    print("We cannot reject the null hypothesis that United took the same number of corners per game as Spurs")
```
